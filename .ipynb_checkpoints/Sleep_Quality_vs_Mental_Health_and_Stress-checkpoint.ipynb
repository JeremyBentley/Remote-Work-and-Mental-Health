```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = "./Impact_of_Remote_Work_on_Mental_Health.csv"
```

```python
mentalHealthDf = pd.read_csv(file)
mentalHealthDf.head()
```

```python
mentalHealthDf.columns
```

```python
mentalHealthDf['Mental_Health_Condition'] = mentalHealthDf['Mental_Health_Condition'].fillna('Healthy')

mentalHealthDf
```

```python
mentalHealthDfReduced = mentalHealthDf[['Sleep_Quality', 'Stress_Level', 'Mental_Health_Condition']]
```

```python
mentalHealthDfHealth = pd.get_dummies(mentalHealthDf['Mental_Health_Condition'])
```

```python
mentalHealthStressSleepandCond = pd.concat([mentalHealthDfReduced, mentalHealthDfHealth], axis = 1)

mentalHealthStressSleepandCond
```

```python
sleepDf = mentalHealthStressSleepandCond.groupby(['Stress_Level', 'Sleep_Quality']).sum(numeric_only=True).reset_index()

sleepDfbyCond = sleepDf[['Stress_Level', 'Sleep_Quality', 'Anxiety', 'Burnout', 'Depression', 'Healthy']]

sleepDfbyCond['Total'] = sleepDfbyCond[['Anxiety', 'Burnout', 'Depression', 'Healthy']].sum(axis=1)

sleepDfbyCond
```
