```python
remote_data.drop(columns = ['Productivity_Change' , 'Physical_Activity', 'Satisfaction_with_Remote_Work'], axis = 1, inplace = True)
```

```python
print(remote_data.columns)
```

```python
remote_data[['Mental_Health_Condition']] = remote_data[['Mental_Health_Condition']].fillna('Not Available')
```

```python
remote_data.isnull().sum()
```

```python
remote_data
```

```python
# Group by region and count the stress levels

region_stress_count_df = remote_data.groupby('Region')['Stress_Level'].size().reset_index(name = 'stress_level_Count')
region_stress_count_df
```

```python
# Group by 'Industry' and calculate the average hours worked per week

Hours_Per_Week = pd.DataFrame(remote_data)
Hours_Per_Week = Hours_Per_Week.groupby('Industry')['Hours_Worked_Per_Week'].mean().reset_index()
Hours_Per_Week
```

```python
#Renaming the column

Average_Hours_Week_df = Hours_Per_Week.rename(columns={'Hours_Worked_Per_Week': 'Average_Hours_Week'})
Average_Hours_Week_df
```

```python
# # Sort the DataFrame from highest to lowest average hours

# Average_Hours_Week_sort = Average_Hours_Week.sort_values(by = 'Average_Hours_Week' ,ascending = False )#.reset_index()
# Average_Hours_Week_sort
```

```python
# Now group by 'Job_loc' and calculate the mean of the new 'Mental_Health_Score'
work_loc_stats = remote_data.groupby('Work_Location')['Mental_Health_Condition'].size().reset_index(name = 'mental_health_count')

# Sort the results to find the worst job loc
awful_work_loc_df = work_loc_stats.sort_values(by='Work_Location')

# Print the job loc with the worst mental health condition
#print(worst_job_loc)
awful_work_loc_df
```
