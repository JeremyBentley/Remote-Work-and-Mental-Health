```python
import os
import pandas as pd
```

```python
file = "./Impact_of_Remote_Work_on_Mental_Health.csv"

mentalHealthDf = pd.read_csv(file)
mentalHealthDf.head()
```

```python
mentalHealthDf.columns
```

```python
mentalHealthDf['Mental_Health_Condition'] = mentalHealthDf['Mental_Health_Condition'].fillna('Healthy')

mentalHealthDf
```

```python
#group ages and add column to dataframe
bins = [20, 30, 40, 50, 60, 70]
labels = ['20-29', '30-39', '40-49', '50-59', '60-69']

mentalHealthDf['ageGroup'] = pd.cut(mentalHealthDf['Age'], bins=bins, labels=labels, right = False)

mentalHealthDf
```

```python
#setup variables for mental health conditions and concatenate
mentalHealthbyCond = pd.get_dummies(mentalHealthDf['Mental_Health_Condition'])

mentalHealthDf = pd.concat([mentalHealthDf, mentalHealthbyCond], axis=1)
```

```python
#setup separate dataframes
hybrid_df = mentalHealthDf[mentalHealthDf['Work_Location'] == 'Hybrid']
onsite_df = mentalHealthDf[mentalHealthDf['Work_Location'] == 'Onsite']
remote_df = mentalHealthDf[mentalHealthDf['Work_Location'] == 'Remote']
```

```python
#group each dataframe by agegroup
hybrid_grouped = hybrid_df.groupby('ageGroup').sum(numeric_only=True).reset_index()
onsite_grouped = onsite_df.groupby('ageGroup').sum(numeric_only=True).reset_index()
remote_grouped = remote_df.groupby('ageGroup').sum(numeric_only=True).reset_index()
```

```python
#reduce columns 
hybrid_result = hybrid_grouped[['ageGroup', 'Anxiety', 'Burnout', 'Depression', 'Healthy']]
onsite_result = onsite_grouped[['ageGroup', 'Anxiety', 'Burnout', 'Depression', 'Healthy']]
remote_result = remote_grouped[['ageGroup', 'Anxiety', 'Burnout', 'Depression', 'Healthy']]
```

```python
hybrid_result['Total'] = hybrid_result[['Anxiety', 'Burnout', 'Depression', 'Healthy']].sum(axis=1)
onsite_result['Total'] = onsite_result[['Anxiety', 'Burnout', 'Depression', 'Healthy']].sum(axis=1)
remote_result['Total'] = remote_result[['Anxiety', 'Burnout', 'Depression', 'Healthy']].sum(axis=1)
```

```python
print("Hybrid Work Location DataFrame:")
hybrid_result
```

```python
print("Onsite Work Location DataFrame:")
onsite_result
```

```python
print("Remote Work Location DataFrame:")
remote_result
```

```python
mentalByLocation = mentalHealthDf.groupby('Work_Location').sum(numeric_only=True).reset_index()

mentalByLocation = mentalByLocation[['Work_Location', 'Anxiety', 'Burnout', 'Depression', 'Healthy']]

mentalByLocation
```

```python
mentalByLocation['Total'] = mentalByLocation[['Anxiety', 'Burnout', 'Depression', 'Healthy']].sum(axis=1)

mentalByLocation
```

```python
mentalByLocation['Unhealthy'] = mentalByLocation[['Anxiety', 'Burnout', 'Depression']].sum(axis=1)

mentalByLocation = mentalByLocation.drop(columns=['Anxiety', 'Burnout', 'Depression'])

mentalByLocation
```
