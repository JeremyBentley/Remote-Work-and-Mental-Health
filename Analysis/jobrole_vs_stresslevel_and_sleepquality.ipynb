```python
jobrole_stress_sleep_df1 = remote_data[['Job_Role','Stress_Level','Sleep_Quality']]
jobrole_stress_sleep_df1
```

```python
jobrole_stress_sleep_df2 = pd.get_dummies(jobrole_stress_sleep_df1['Sleep_Quality'])
```

```python
jobrole_stress_sleep_df3 = pd.concat([jobrole_stress_sleep_df1, jobrole_stress_sleep_df2],axis = 1)
jobrole_stress_sleep_df3
```

```python
jobrole_stress_sleep_df4 = jobrole_stress_sleep_df3.groupby(['Job_Role','Stress_Level']).sum(numeric_only=True)
jobrole_stress_sleep_df4
```

```python
jobrole_stress_sleep_df5 = jobrole_stress_sleep_df3.groupby(['Job_Role','Stress_Level']).sum(numeric_only=True).reset_index()
jobrole_stress_sleep_df5
```

```python
jobrole_stress_sleep_total= jobrole_stress_sleep_df5[['Job_Role','Stress_Level','Good','Average','Poor']]
jobrole_stresslevel_sleepquality_total = jobrole_stress_sleep_total.copy()
jobrole_stresslevel_sleepquality_total.loc[:,'Total'] = jobrole_stress_sleep_total[['Good','Average','Poor']].sum(axis=1)
```

```python
jobrole_stresslevel_sleepquality_total.to_csv("./jobrole.csv", index=False)
```
