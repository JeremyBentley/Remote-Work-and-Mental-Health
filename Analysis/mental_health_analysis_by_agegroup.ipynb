```python
import os
import pandas as pd
```

```python
file = "./Impact_of_Remote_Work_on_Mental_Health.csv"
```

```python
mentalHealthDf = pd.read_csv(file)
mentalHealthDf.head()
```

```python
mentalHealthDf.columns
```

```python
mentalHealthDf['Mental_Health_Condition'] = mentalHealthDf['Mental_Health_Condition'].fillna('Healthy')

mentalHealthDf
```

```python
mentalHealthByAge = mentalHealthDf[['Age', 'Mental_Health_Condition']]
```

```python
bins = [20, 30, 40, 50, 60, 70]
labels = ['20-29', '30-39', '40-49', '50-59', '60-69']

mentalHealthByAge['ageGroup'] = pd.cut(mentalHealthByAge['Age'], bins=bins, labels=labels, right = False)

mentalHealthByAge
```

```python
mentalHealthbyCond = pd.get_dummies(mentalHealthByAge['Mental_Health_Condition'])

df_updated = pd.concat([mentalHealthByAge, mentalHealthbyCond], axis=1)

df_updated
```

```python
mentalHealthDfShortened = df_updated.groupby('ageGroup').sum(numeric_only=True).reset_index()

mentalHealthDfShortened
```

```python
result_df = mentalHealthDfShortened[['ageGroup', 'Anxiety', 'Burnout', 'Depression', 'Healthy']]

result_df['Total'] = result_df[['Anxiety', 'Burnout', 'Depression', 'Healthy']].sum(axis=1)

result_df
```

```python
result_df['Unhealthy'] = result_df[['Anxiety', 'Burnout', 'Depression']].sum(axis=1)

result_df = result_df.drop(columns=['Anxiety', 'Burnout', 'Depression'])

result_df
```

```python
result_df = result_df[['ageGroup', 'Healthy', 'Unhealthy', 'Total']]

result_df
```

```python
total_row = result_df[['Healthy', 'Unhealthy', 'Total']].sum()

total_df = pd.DataFrame(data={'ageGroup': ['Total'], 
                                   'Healthy': [total_row['Healthy']], 
                                   'Unhealthy': [total_row['Unhealthy']], 
                                   'Total': [total_row['Total']]})

result_df = pd.concat([result_df, total_df], ignore_index=True)

result_df
```
