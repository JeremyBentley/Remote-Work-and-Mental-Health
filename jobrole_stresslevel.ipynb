```python
# Group by 'Industry' and calculate the average hours worked per week

Hours_Per_Week = pd.DataFrame(remote_data)
Hours_Per_Week = Hours_Per_Week.groupby('Stress_Level')['Mental_Health_Condition','Sleep_Quality'].mean().reset_index()
Hours_Per_Week
```

```python
#Renaming the column

Average_Hours_Week_df = Hours_Per_Week.rename(columns={'Hours_Worked_Per_Week': 'Average_Hours_Week'})
Average_Hours_Week_df
```
