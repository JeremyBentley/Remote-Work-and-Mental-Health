```python
work_Location_mentalhealth_df1 = remote_data[['Work_Location','Mental_Health_Condition']]
work_Location_mentalhealth_df1
```

```python
work_Location_mentalhealth_df2 = pd.get_dummies(work_Location_mentalhealth_df1['Mental_Health_Condition'])
```

```python
work_Location_mentalhealth_df3 = pd.concat([work_Location_mentalhealth_df1, work_Location_mentalhealth_df2],axis = 1)
work_Location_mentalhealth_df3
```

```python
work_Location_mentalhealth_df4 = work_Location_mentalhealth_df3.groupby(['Work_Location']).sum(numeric_only=True).reset_index()
work_Location_mentalhealth_df4
```

```python
work_Location_mentalhealth = work_Location_mentalhealth_df4[['Work_Location','Anxiety','Burnout','Depression']]
work_Location_mentalhealth_total = work_Location_mentalhealth.copy()
work_Location_mentalhealth_total.loc[:,'Total'] = work_Location_mentalhealth[['Anxiety','Burnout','Depression']].sum(axis=1)
```
